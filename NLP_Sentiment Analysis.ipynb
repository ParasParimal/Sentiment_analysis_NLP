```python
#naming the columns and reading the dataset again
column_names=['target', 'id', 'date', 'flag', 'user', 'text']
data = pd.read_csv(r'C:\Users\HP\OneDrive\Documents\NLP Project\Train_dataset.csv', encoding='latin1', names=column_names, encoding_errors='ignore')
data.shape
```

```python
# convert the target "4" to "1"
data.replace({'target':{4:1}},inplace=True)
data['target'].value_counts()
```

```python
data.sample(10)
```

```python
# O means negative tweet and if 1 then it is positive tweet
# stemming
port_stem=PorterStemmer() # it reduces the word to its root word
def stemming(content):
    stemmed_content=re.sub('[^a-zA-Z]',' ', content)
    stemmed_content=stemmed_content.lower()
    stemmed_content=stemmed_content.split()
    stemmed_content=[port_stem.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
    stemmed_content=' '.join(stemmed_content)
    return stemmed_content
```

```python
data['stemmed_content']=data['text'].apply(stemming)
```

```python
# separating the data and the label
X=data['stemmed_content'].values
Y=data['target'].values
```

```python
print(X)
```

```python
print(Y)
```

```python
# Splitting the data to training and test data

# X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, stratify=Y, random_state=2)
# print(X.shape, X_train.shape, X_test.shape)
#data = data[['stemmed_content', 'target']]

# Step 4: Preprocess text (optional: lowercase, remove nulls)
#data['stemmed_content'] = data['stemmed_content'].astype(str).str.lower().fillna('')

# Step 5: Split data into train/test
X_train, X_test, y_train, y_test = train_test_split(
    X,Y, test_size=0.2, random_state=42)
```

```python
print(X_test)
```

```python
# Converting the textual data to numerical data by appling TF-IDF on the raw text
vectorizer = TfidfVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

```

```python
print(X_train_vec)
```

```python
print(X_test_vec)
```

```python
X_train_vec.shape
```

```python
X_test_vec.shape
```

```python
#Training the Machine Learning
#Logistic Regression
model=LogisticRegression(max_iter=5000)
model.fit(X_train_vec, y_train)
```

```python
#Evaluation of the model
X_train_prediction = model.predict(X_train_vec)
training_data_accuracy = accuracy_score(y_train, X_train_prediction)
print('Accuracy score on the training data:', training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test_vec)
test_data_accuracy = accuracy_score(y_test, X_test_prediction)
print('Accuracy score on the test data:', test_data_accuracy)
```
